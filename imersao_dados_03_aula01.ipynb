Para que seja possivel utilizar o método query nas colunas g-* e c-* é preciso manipular os nomes para que seja retirado o hífen


```python
# 06
cols = {'g-0':'g0'}

dados.rename(columns=cols)
```

```python
# 07
import matplotlib.pyplot as plt

fig1, ax1 = plt.subplots(figsize = (10,6))

_dados = dados['tratamento'].value_counts(normalize = True)
ax1.pie(_dados, labels = ['Com droga', 'Com controle'], autopct = '%1.1f%%', colors = ['darkblue', 'darkorchid'])
plt.suptitle('Aula 1 Imersão Dados')
plt.title('Proporção de tratamentos')
ax1.legend()
plt.show()
```
